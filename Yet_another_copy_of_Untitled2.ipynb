###**Importing modules:**

like matplotlib, random, and math.

```python
# !pip install ipypwidget
from ipywidgets import HTML
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
import math
```

###**Creating class:**
In this cell a class 'Circle' is created with attributes x,y,radius,color,velocity. Also the functions move, collides and after collision are created.

###**working of functions:**
*  **Function move**-> The position of the circle is being updated according to its velocity. the 'if' statements handle the collision of the ball with walls
*  **Function collides**-> this funcition check if the  circles collied with each othere or not.
*  **Function after collision**-> this function is the after math of collision if the circles collied with each othere then the volcity of the circles are inter changed


```python
class Circle:
    def __init__(self, x, y, radius, color, velocity):
        self.x = x
        self.y = y
        self.radius = radius
        self.color = color
        self.velocity = velocity

    def move(self):
        self.x += self.velocity[0]
        self.y += self.velocity[1]


        if self.x - self.radius < 0 or self.x + self.radius > 100:
            self.velocity = (-self.velocity[0], self.velocity[1])

        if self.y - self.radius < 0 or self.y + self.radius > 100:
            self.velocity = (self.velocity[0], -self.velocity[1])

    def collides(self, other_circle):
        distance = math.sqrt((self.x - other_circle.x)**2 + (self.y - other_circle.y)**2)
        return distance < self.radius + other_circle.radius

    def after_collision(self, other_circle):

        self.velocity, other_circle.velocity = other_circle.velocity, self.velocity
```

###**Update function:**
In this function the simulation is being created. This function moves the circle frist then check if there is a colllision. after this,  remove the patch  of circle and creat a new one which creats an animation.


```python
def update(frame):
    for i, circle in enumerate(circles):
        circle.move()


        for other_circle in circles[i + 1:]:
            if circle.collides(other_circle):
                circle.after_collision(other_circle)


    for circle_patch in ax.patches:
        circle_patch.remove()


    for circle in circles:
        circle_patch = plt.Circle((circle.x, circle.y), circle.radius, color=circle.color)
        ax.add_patch(circle_patch)

    return ax.patches
```

###**Creating circles:**
list comprihension is used to create circles


```python
circles = [
    Circle(random.uniform(0, 100), random.uniform(0, 100), 3 , random.choice(['red','blue','green','pink','brown']), (random.uniform(-2, 2), random.uniform(-2, 2)))
    for _ in range(10)
]
```

###***Ploting graph:***
In this cell the graph is plotted and Funcanimation is used to creat animation


```python
fig, ax = plt.subplots()
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)


ani = FuncAnimation(fig, update, frames=100, interval=100, blit=True)

plt.close()
```
