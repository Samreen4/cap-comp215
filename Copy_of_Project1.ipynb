importing the modules we need

```python
from ipywidgets import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
import numpy as np
import math

```

creating the class of circle with functions move and collide. move function change the positon of the circles and the function collides detect the collision of the circle.

```python
class Circle:
    def __init__(self, x, y, radius, velocity=(0, 0)):
        self.x = x
        self.y = y
        self.radius = radius
        self.velocity = velocity


    def move(self, dt):
        self.x += self.velocity[0] * dt
        self.y += self.velocity[1] * dt


    def collides(self, other_circle):
        distance = math.sqrt((self.x - other_circle.x)**2 + (self.y - other_circle.y)**2)
        return distance < self.radius + other_circle.radius

```

creating the circles , figure , axis,and patches of circle. also limit of the axis

```python
# create the circles
circle1 = Circle(0, 0, 1, (1, 0))
circle2 = Circle(0, 1, 1, (0, -1))


# create the figure and axis
fig, ax = plt.subplots()


# create the patches for the circles
circle1_patch = patches.Circle((circle1.x, circle1.y), circle1.radius, color='r')
circle2_patch = patches.Circle((circle2.x, circle2.y), circle2.radius, color='b')


# add the patches to the axis
ax.add_patch(circle1_patch)
ax.add_patch(circle2_patch)


# set the limits of the axis
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
```

creating a function update which updates the frame. if the circle collides changes the velocity

```python
def update(data):

    # move the circles
    circle1.move(data)
    circle2.move(data)


    # detect collision and reverse velocities
    if circle1.collides(circle2):
        circle1.velocity = (-circle1.velocity[0], -circle1.velocity[1])
        circle2.velocity = (-circle2.velocity[0], -circle2.velocity[1])


     #update the positions of the patches
    circle1_patch = patches.Circle((circle1.x, circle1.y), circle1.radius, color='r')
    circle2_patch = patches.Circle((circle2.x, circle2.y), circle2.radius, color='b')


    return circle1_patch, circle2_patch

```
